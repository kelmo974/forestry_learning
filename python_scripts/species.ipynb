```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file_path = 'data/v9-5_2024-10_Natl_MasterTreeSpeciesList.xlsx'
```

```python
df_species = pd.read_excel(file_path)
```

```python
df_species.head()
```

```python
keep_cols = ['FIA Code', 'Common Name', 'Genus', 'Species', 'SCIENTIFIC_NAME', 'List Status']
df_species = df_species[keep_cols]
```

```python
df_species = df_species.rename(columns={
    'FIA Code': 'spcd',
    'Common Name': 'common_name',
    'Genus': 'genus',
    'Species': 'species',
    'SCIENTIFIC_NAME': 'scientific_name',
    'List Status': 'list_status'
})
```

```python
df_species.head()
```

```python
engine = create_engine('postgresql://kellen@localhost:5432/forestry_research')
```

```python
df_species.to_sql('species_list', engine, schema='raw_data', if_exists='append', index=False)
```
